```python
from typing import List, Dict, Tuple, Optional, Union, Any

from prompt_template_collection import PromptTemplate
from doraemon import Doraemon
from relaxed_fda import RelaxedFDA

task_name = "FEVER"
logger = Doraemon.get_logger(name=__name__, logfile="relaxed_FDA_on_fever.log")
file_path='/kaggle/input/building-sot-on-fever/sots_df.pkl'

grouped_data = RelaxedFDA.prepare_dataset(file_path=file_path,mode='r_s', logger=logger)

encoder=RelaxedFDA.get_encoder()

# Demo promts question, corect, and wrong 
D: List[Dict] = PromptTemplate.sot_construct_inter_fever()
logger.info(len(D))

op_system_prompt = RelaxedFDA.get_optimize_system_prompt(task_name)
logger.info(op_system_prompt)
```

```python
import pandas as pd

result_pd: pd.DataFrame = RelaxedFDA.evaluate(
    grouped_data,
    logger,
    op_system_prompt,
    encoder,
    D,
    enable_logger_rs=False
)

result_pd = result_pd.dropna()
result_pd.to_pickle('result_pd.pkl')  # Saves to a pickle file
```

```python
logger.info(f"Accuracy on {task_name} Dataset:{RelaxedFDA.calculate_accuracy(result_pd):.2%}")
```
