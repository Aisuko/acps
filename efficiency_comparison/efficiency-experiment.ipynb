```python
import matplotlib.pyplot as plt

# Data for StrategyQA
tokens = [0, 100, 200, 300, 400, 500]
acc_cp_strategyqa   = [None, 66.67, 72.50, 75.00, 79.17, 83.11]
acc_ours_strategyqa = [None, 78.00, 80.01, 81.13, 82.11, 84.10]

# Data for HotpotQA
acc_cp_hotpotqa   = [None, 58.13, 59.10, 60.70, 61.75, 62.32]
acc_ours_hotpotqa = [None, 59.04, 60.82, 61.31, 62.10, 63.21]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# StrategyQA plot (left)
axes[0].plot(tokens, acc_cp_strategyqa, marker="^", linestyle="-", color="purple", label="CP")
axes[0].plot(tokens, acc_ours_strategyqa, marker="*", linestyle="-", color="orange", label="Ours")
axes[0].set_xlabel("Number of Tokens")
axes[0].set_ylabel("Accuracy (%)")
axes[0].set_title("Tokens vs Accuracy: CP and Ours (StrategyQA)")
axes[0].legend()
axes[0].grid(True)
axes[0].set_xticks(tokens)

# HotpotQA plot (right)
axes[1].plot(tokens, acc_cp_hotpotqa, marker="^", linestyle="-", color="purple", label="CP")
axes[1].plot(tokens, acc_ours_hotpotqa, marker="*", linestyle="-", color="orange", label="Ours")
axes[1].set_xlabel("Number of Tokens")
axes[1].set_ylabel("Accuracy (%)")
axes[1].set_title("Tokens vs Accuracy: CP and Ours (HotpotQA)")
axes[1].legend()
axes[1].grid(True)
axes[1].set_xticks(tokens)

plt.tight_layout()
plt.savefig("tokens_vs_accuracy_combined.png", dpi=300, bbox_inches="tight")
plt.show()
```
